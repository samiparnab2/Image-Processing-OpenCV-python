```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
#getting path to the image
path=input("enter the absolute path to the image")
#getting the matrix of RGB intensities using cv2.imread()
imgBGR=cv2.imread(path)
#converting from BGR to RGB
imgRGB=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
#converting from BGR to GrayScale
imgGray=cv2.cvtColor(imgBGR,cv2.COLOR_BGR2GRAY)
#creating a RGB image matrix which have intensity values only for 1 channel and 0 for the remaining two using list comprehension 
imgR=[[ [imgRGB[i,j,k] if k==0 else 0 for k in range(0,3)] for j in range(0,imgRGB.shape[1])]for i in range(0,imgRGB.shape[0])]
imgG=[[ [imgRGB[i,j,k] if k==1 else 0 for k in range(0,3)] for j in range(0,imgRGB.shape[1])]for i in range(0,imgRGB.shape[0])]
imgB=[[ [imgRGB[i,j,k] if k==2 else 0 for k in range(0,3)] for j in range(0,imgRGB.shape[1])]for i in range(0,imgRGB.shape[0])]
#converting form list to np.array
imgR=np.array(imgR)
imgG=np.array(imgG)
imgB=np.array(imgB)

#setting figure size of the image plot and creating subplot for each of the images(color image)
plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.imshow(imgRGB)
plt.title("colorImage")

plt.subplot(1,4,2)
plt.imshow(imgR)
plt.title("RedImage")

plt.subplot(1,4,3)
plt.imshow(imgG)
plt.title("GreenImage")

plt.subplot(1,4,4)
plt.imshow(imgB)
plt.title("blueImage")
plt.show()

#setting figure size of the image plot and creating subplot for each of the images(GrayScale image)
plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.imshow(imgGray,cmap='gray')
plt.title("ColorImage(grayscale intensity)")

plt.subplot(1,4,2)
plt.imshow(imgRGB[:,:,0],cmap='gray')
plt.title("RedImage(grayscale intensity)")

plt.subplot(1,4,3)
plt.imshow(imgRGB[:,:,1],cmap='gray')
plt.title("GreenImage(grayscale intensity)")

plt.subplot(1,4,4)
plt.imshow(imgRGB[:,:,2],cmap='gray')
plt.title("BlueImage(grayscale intensity)")
plt.show()
```
