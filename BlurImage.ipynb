```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
path=input("entyer absolute path to the file")
#reading grascale values
imgGray=cv2.imread(path,0)
row,col=imgGray.shape
imgPadded=np.zeros((row+2,col+2),dtype=np.uint8)
imgPadded[1:row+1,1:col+1]=imgGray
#kernel whichh is used
kernel=[[1,1,1],
        [1,1,1],
        [1,1,1]]
kernel=np.array(kernel)
ksum=np.sum(kernel)
imgBlur=np.zeros((row,col),dtype=np.uint8)
#convolution operation
for i in range(0,row):
    for j in range(0,col):
        imgBlur[i,j]=(np.sum(imgPadded[i:i+3,j:j+3]*kernel))/ksum
#plotting images
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(imgGray,cmap='gray')
plt.title("orginal")
plt.subplot(1,2,2)
plt.imshow(imgBlur,cmap='gray')
plt.title("blured")
plt.show()
```
