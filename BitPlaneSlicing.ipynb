```python
import cv2
import matplotlib.pyplot as plt
path=input("enter absolute path of the image")
#reading the image in GrayScale
imgGray=cv2.imread(path,0)
slicedImg=[]
bitVal=1
plt.figure(figsize=(20,20))
for i in range(0,8):
    #Taking the ith bit value and storing it on new array
    slicedImg.append([])
    slicedImg[len(slicedImg)-1]=imgGray & bitVal
    bitVal=bitVal<<1
    #plotting the image
    plt.subplot(2,4,i+1)
    plt.imshow(slicedImg[i],cmap='gray')
    plt.title("BitPlane"+str(i))
plt.show()
```
