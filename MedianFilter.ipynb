```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
path=input("entyer absolute path to the file")
#reading grascale values
imgGray=cv2.imread(path,0)
row,col=imgGray.shape
wx=5
wy=5
totalElem=wx*wy
incy=wy//2
incx=wx//2
imgPadded=np.zeros((row+(incy*2),col+(incx*2)),dtype=np.uint8)
imgPadded[incy:imgPadded.shape[0]-incy,incx:imgPadded.shape[1]-incx]=imgGray
imgNew=np.zeros((row,col),dtype=np.uint8)
for i in range(0,row):
    for j in range(0,col):
        imgNew[i,j]=np.sort(imgPadded[i:i+wy,j:j+wx],None)[totalElem//2]
#plotting images
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(imgGray,cmap='gray')
plt.title("orginal")
plt.subplot(1,2,2)
plt.imshow(imgNew,cmap='gray')
plt.title("Changed")
plt.show()
```
