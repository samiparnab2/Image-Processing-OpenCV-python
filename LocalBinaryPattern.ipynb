```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
path=input("enter absolute path of image")
#reading image in grascale
imgGray=cv2.imread(path,0)
row,col=imgGray.shape
#padding image with zeros on the boundaries
padImg=np.zeros((row+2,col+2),dtype=np.uint8)
padImg[1:row+1,1:col+1]=imgGray
#taking the wight matrix for calculating the resultant intensity
weight=np.array([[128,64,32],[1,0,16],[2,4,8]])
#empty matrix that will store the lbp
lbpImg=np.zeros((row,col),dtype=np.uint8)
for i in range(0,row):
    for j in range(0,col):
        val=np.greater(padImg[i:i+3,j:j+3],padImg[i+1,j+1])&1  #comparing the surrounding 8 values and performing bitwise AND to get 1 where the value is greater and 0 otherwise
        lbpImg[i,j]=np.sum(val*weight)    #just multiplying each weights to the corresponding binary pattern and getting the sum

#plotting the images
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(lbpImg,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(imgGray,cmap='gray')
plt.show()
```
